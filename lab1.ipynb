```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def function(x):
    return x**2+5
```

```python
x1 = np.linspace(-1,1,1000)
x2 = np.linspace(-6,6,1000)
x3 = np.linspace(0,5,1000)
y1 = function(x1)
y2 = function(x2)
y3 = function(x3)
```

```python
plt.title("Wykres 1") 
plt.xlabel("x") 
plt.ylabel("y") 
plt.plot(x1,y1) 
plt.show()
```

```python
plt.title("Wykres 1") 
plt.xlabel("x") 
plt.ylabel("y") 
plt.plot(x2,y2) 
plt.show()
```

```python
plt.title("Wykres 1") 
plt.xlabel("x") 
plt.ylabel("y") 
plt.plot(x3,y3) 
plt.show()
```
