```python
import pandas as pd
import numpy as np

df = pd.read_csv("/home/szymon/Documents/AIR_5/AiBD/laby_2022/tier/OriginalData/tb.csv")
df
```

```python
#usunięcie wartości Nan oraz kolumn, które nie mają sensu(zakres wieku od 0 do 14, gdy obecne są dwie kolumny od 0 do 4 i od 5 do 14 )
df = df.dropna()
df = df.drop(columns=['new_sp'])
df = df.drop(columns=['new_sp_m014'])
df = df.drop(columns=['new_sp_f014'])
df
```

```python
#zmiana nazw kolumn
df.rename(columns={'new_sp_m04':'m04', 
                   'new_sp_m514':'m514',      
                   'new_sp_m1524':'m1524',
                   'new_sp_m2534':'m2534',
                   'new_sp_m3544':'m3522',
                   'new_sp_m4554':'m4554',
                   'new_sp_m5564':'m5564',
                   'new_sp_m65':'m65',
                   'new_sp_mu':'mu',
                   'new_sp_mu':'fu',       
                   'new_sp_f04':'f04',
                   'new_sp_f514':'f514',
                   'new_sp_f1524':'f1524',
                   'new_sp_f2534':'f2534',      
                   'new_sp_f3544':'f3544',
                   'new_sp_f4554':'f4554',
                   'new_sp_f5564':'f5564',
                   'new_sp_f65':'f65', 
                   'new_sp_fu':'fu', 
                   }, inplace=True)
df
```

```python
#porządkowanie
df = pd.melt(df, id_vars=['iso2', 'year'], value_vars=list(df.columns)[2:],var_name='column', value_name='cases')
df['cases'] = df['cases'].astype(int)
df['sex'] = df['column'].str[0]
df['age'] = df['column'].str[1:].map({
    '04': '0-4',
    '514': '5-14',
    '1524': '15-24',
    '2534': '25-34',
    '3544': '35-44',
    '4554': '45-54',
    '5564': '55-64',
    '65': '65+',
  
})
df.to_csv('../AnalysisData/tb_tidy.csv')
df
```

```python
#porządkowanie względem płci - mężczyźni
df_male = df.groupby(df.sex).get_group('m')
df_male_final = pd.DataFrame(df_male.groupby(['age', 'cases']).size().groupby(level=0).max()).reset_index()
df_male_final
```

```python
#porządkowanie względem płci - kobiety
df_female = df.groupby(df.sex).get_group('f')
df_female_final = pd.DataFrame(df_female.groupby(['age', 'cases']).size().groupby(level=0).max()).reset_index()
df_female_final
```
